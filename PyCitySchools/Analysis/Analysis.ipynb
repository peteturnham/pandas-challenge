## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
math_pass= df[df['math_score']>=70].groupby(['school_name']).size()
```

```python
math_p= (math_pass/len(df['student_name'])*1000)
```

```python
read_pass=df[df['reading_score']>=70].groupby(['school_name']).size()
```

```python
read_p=(read_pass/len(df['student_name'])*1000)
```

```python
rm_p= (read_p+math_p)/2
```

```python
math=pd.DataFrame(math_p)
read=pd.DataFrame(read_p)
read_math=pd.DataFrame(rm_p)
merged_math_read=pd.merge(math, read,on="school_name")
sum_score=pd.merge(merged_math_read, read_math,on="school_name")
sum_score.head()
combined= pd.merge(school_summary,sum_score,on="school_name")
combined_rename=combined.rename(columns={"0_x":"Percent Passing Math",
                                         "0_y":"Percent Passing Reading",
                                         "0":"Percent of Both"})
```
